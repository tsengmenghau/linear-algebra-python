```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定義使用者喜好向量
alice = np.array([4, 1])
bob = np.array([1, 5])
carol = np.array([3, 2])
```

```python
# 畫出 quiver 圖
plt.quiver(0, 0, alice[0], alice[1], angles='xy', scale_units='xy', scale=1, color='r', label='Alice')
plt.quiver(0, 0, bob[0], bob[1], angles='xy', scale_units='xy', scale=1, color='g', label='Bob')
plt.quiver(0, 0, carol[0], carol[1], angles='xy', scale_units='xy', scale=1, color='b', label='Carol')

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.gca().set_aspect('equal')
plt.grid()
plt.legend()
plt.title("User Preferences Vectors")
plt.show()
```

```python
# 計算內積（相似度）
sim_alice_bob = np.dot(alice, bob)
sim_alice_carol = np.dot(alice, carol)

print("Alice · Bob =", sim_alice_bob)
print("Alice · Carol =", sim_alice_carol)
```
