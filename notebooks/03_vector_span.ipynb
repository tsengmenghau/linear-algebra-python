```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([2, 1])
v2 = np.array([-1, 3])

# 建立多組線性組合
a_values = np.linspace(-2, 2, 5)
b_values = np.linspace(-2, 2, 5)

plt.figure(figsize=(6,6))

for a in a_values:
    for b in b_values:
        v = a * v1 + b * v2
        plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)

# 原始向量
plt.quiver(0, 0, v1[0], v1[1], color='r', angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(0, 0, v2[0], v2[1], color='b', angles='xy', scale_units='xy', scale=1, label='v2')

```
