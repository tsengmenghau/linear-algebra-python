```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
v1 = np.array([2, 1])
```

```python
v2 = np.array([-1,3])
```

```python
v3 = v1 + v2
```

```python
plt.quiver(0,0, v1[0], v2[1], color='r', angles ='xy', scale_units='xy', scale=1, label='v1')
```

```python
plt.quiver(0,0, v2[0], v2[1], color='b', angles='xy', scale_units='xy', scale=1, label='v2')
```

```python
plt.quiver(0, 0, v3[0], v3[1], color='g', angles='xy', scale_units='xy', scales=1, label='v1+v2')
```

```python
plt.quiver(0, 0, v3[0], v3[1], color='g', angles='xy', scale_units='xy', scale=1, label='v1+v2')
```

```python
plt.xlim(-5,5)
```

```python
plt.ylim(-5,5)
```

```python
plt.gca().set_aspect('equal')
```

```python
plt.grid()
```

```python
plt.title("Vector Addition")
```

```python
plt.legend()
```

```python
plt.show()
```

```python
plt.quiver(0, 0, v1[0], v1[1], color='r', angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(0, 0, v2[0], v2[1], color='b', angles='xy', scale_units='xy', scale=1, label='v2')
plt.quiver(0, 0, v3[0], v3[1], color='g', angles='xy', scale_units='xy', scale=1, label='v1+v2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal')
plt.grid()
plt.title("Vector Addition")
plt.legend()
plt.show()
```
