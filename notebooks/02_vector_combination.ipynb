```python
import numpy as np
import matplotlib.pyplot as plt

# 原始向量
v1 = np.array([2, 1])
v2 = np.array([-1, 3])

# 線性組合：v3 = a*v1 + b*v2
a, b = 1, 2
v3 = a * v1 + b * v2

# 畫圖
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='g', label='v3 = a*v1 + b*v2')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.gca().set_aspect('equal')
plt.title("Vector Linear Combination")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 原始向量
v1 = np.array([2, 1])
v2 = np.array([-1, 3])

# 線性組合：v3 = a*v1 + b*v2
a, b = 1, 2
v3 = a * v1 + b * v2

# 畫圖
plt.quiver(0, 0, v2[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, v1[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='g', label='v3 = a*v1 + b*v2')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.gca().set_aspect('equal')
plt.title("Vector Linear Combination")
plt.legend()
plt.show()
```

```python
# 測試不同線性組合的效果
combinations = [(1, 2), (2, -1), (0.5, 0.5), (-1, 1)]

plt.figure(figsize=(6,6))

for a, b in combinations:
    v3 = a * v1 + b * v2
    plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, label=f'{a}*v1 + {b}*v2')

# 原始向量一起畫上
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal')
plt.grid()
plt.legend()
plt.title("Various Linear Combinations of v1 and v2")
plt.show()
```
